```julia
using CSV
using SymPy
using Graphs
```

```julia
datos = CSV.read("entrenamiento.data", header = false)
prueba = CSV.read("prueba.data", header = false)
```

```julia
ent = Bool[]
for i = 1:5716
    length(unique(datos[:,i])) >= 2 ? push!(ent, true) : push!(ent, false)
end
datos = datos[ent]
prueba = prueba[ent]
```

```julia
function attributes(k::Int64)
    atributos = []
    for name = names(datos)
        datos[k,:][name] == 1 ? push!(atributos, Sym(String(name))) : push!(atributos, Not(Sym(String(name))))
    end
    atributos
end
```

```julia
attributes(1)
```

```julia
function grafo(clase::Int64)
    negativos = datos[datos[:,size(datos,2)] .!= clase,:]
    aristas = fill([], size(negativos,1))
    
    for columna = names(negativos)
        for i = 1:size(negativos, 1)
            for j = setdiff(1:size(negativos,1), i)
                if negativos[columna][i] == negativos[columna][j] 
                    push!(aristas[i], (j, columna))
                end
            end
        end
    end
    
    aristas
end
```

```julia
arist = grafo(3)
```
